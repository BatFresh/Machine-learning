```python
#Feature Engineering ç‰¹å¾å·¥ç¨‹
#å¤‡é€‰ç‰¹å¾ï¼šPclassï¼ŒSexï¼ŒEmbarked,cabin,Age
#ä½†Ageå’ŒCabinæ˜¯æœ‰ä¸¢å¤±æ•°æ®çš„ç‰¹å¾ï¼Œ
#æŒ‰Cabinæœ‰æ— æ•°æ®ï¼Œå°†è¿™ä¸ªå±žæ€§å¤„ç†æˆYeså’ŒNoä¸¤ç§
#Ageï¼š
#é€šå¸¸é‡åˆ°ç¼ºå€¼çš„æƒ…å†µï¼Œæˆ‘ä»¬ä¼šæœ‰å‡ ç§å¸¸è§çš„å¤„ç†æ–¹å¼

#å¦‚æžœç¼ºå€¼çš„æ ·æœ¬å æ€»æ•°æ¯”ä¾‹æžé«˜ï¼Œæˆ‘ä»¬å¯èƒ½å°±ç›´æŽ¥èˆå¼ƒäº†ï¼Œä½œä¸ºç‰¹å¾åŠ å…¥çš„è¯ï¼Œå¯èƒ½åå€’å¸¦å…¥noiseï¼Œå½±å“æœ€åŽçš„ç»“æžœäº†
#å¦‚æžœç¼ºå€¼çš„æ ·æœ¬é€‚ä¸­ï¼Œè€Œè¯¥å±žæ€§éžè¿žç»­å€¼ç‰¹å¾å±žæ€§(æ¯”å¦‚è¯´ç±»ç›®å±žæ€§)ï¼Œé‚£å°±æŠŠNaNä½œä¸ºä¸€ä¸ªæ–°ç±»åˆ«ï¼ŒåŠ åˆ°ç±»åˆ«ç‰¹å¾ä¸­
#å¦‚æžœç¼ºå€¼çš„æ ·æœ¬é€‚ä¸­ï¼Œè€Œè¯¥å±žæ€§ä¸ºè¿žç»­å€¼ç‰¹å¾å±žæ€§ï¼Œæœ‰æ—¶å€™æˆ‘ä»¬ä¼šè€ƒè™‘ç»™å®šä¸€ä¸ªstep(æ¯”å¦‚è¿™é‡Œçš„ageï¼Œæˆ‘ä»¬å¯ä»¥è€ƒè™‘æ¯éš”2/3å²ä¸ºä¸€ä¸ªæ­¥é•¿)ï¼Œç„¶åŽæŠŠå®ƒç¦»æ•£åŒ–ï¼Œä¹‹åŽæŠŠNaNä½œä¸ºä¸€ä¸ªtypeåŠ åˆ°å±žæ€§ç±»ç›®ä¸­ã€‚
#æœ‰äº›æƒ…å†µä¸‹ï¼Œç¼ºå¤±çš„å€¼ä¸ªæ•°å¹¶ä¸æ˜¯ç‰¹åˆ«å¤šï¼Œé‚£æˆ‘ä»¬ä¹Ÿå¯ä»¥è¯•ç€æ ¹æ®å·²æœ‰çš„å€¼ï¼Œæ‹Ÿåˆä¸€ä¸‹æ•°æ®ï¼Œè¡¥å……ä¸Šã€‚

#å…ˆé‡‡ç”¨æ‹Ÿåˆå¹´é¾„æ•°æ®ï¼Œé‡‡ç”¨Scikit-learnä¸­çš„RandomForestï¼šä¸€ä¸ªç”¨åœ¨åŽŸå§‹æ•°æ®ä¸­åšä¸åŒé‡‡æ ·ï¼Œå»ºç«‹å¤šé¢—DecisionTreeï¼Œå†è¿›è¡Œaverageç­‰ç­‰æ¥é™ä½Žè¿‡æ‹ŸåˆçŽ°è±¡ï¼Œæé«˜ç»“æžœçš„æœºå™¨å­¦ä¹ ç®—æ³•

from sklearn.ensemble import RandomForestRegressor

### ä½¿ç”¨ RandomForestClassifier å¡«è¡¥ç¼ºå¤±çš„å¹´é¾„å±žæ€§

def set_missing_ages(df):
    # æŠŠå·²æœ‰çš„æ•°å€¼åž‹ç‰¹å¾å–å‡ºæ¥ä¸¢è¿›Random Forest Regressorä¸­
    
    age_df = df[['Age','Fare','Parch','SibSp','Pclass']]
    
    # ä¹˜å®¢åˆ†æˆå·²çŸ¥å¹´é¾„å’ŒæœªçŸ¥å¹´é¾„ä¸¤éƒ¨åˆ†
    
    known_age = age_df[age_df.Age.notnull()].as_matrix()
    unknown_age = age_df[age_df.Age.isnull()].as_matrix()
    
    # yå³ç›®æ ‡å¹´é¾„
    y = known_age[:,0]
    
    # Xå³ç‰¹å¾å±žæ€§å€¼
    X = known_age[:, 1:]
    
    # fitåˆ°RandomForestRegressorä¹‹ä¸­
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
    rfr.fit(X, y)
        
    # ç”¨å¾—åˆ°çš„æ¨¡åž‹è¿›è¡ŒæœªçŸ¥å¹´é¾„ç»“æžœé¢„æµ‹
    predictedAges = rfr.predict(unknown_age[:,1::]) 
    
    # ç”¨å¾—åˆ°çš„é¢„æµ‹ç»“æžœå¡«è¡¥åŽŸç¼ºå¤±æ•°æ®
    df.loc[ (df.Age.isnull()), 'Age' ] = predictedAges 

    return df, rfr
        
def set_Cabin_type(df):
    df.loc[(df.Cabin.notnull()),'Cabin'] = 'YES'
    df.loc[(df.Cabin.isnull()),'Cabin'] = 'NO'
    
    return df

data_train,rfr = set_missing_ages(data_train)
data_train = set_Cabin_type(data_train)

#ä»¥ä¸Šè¿è¡Œä¸€æ¬¡åŽï¼Œdata_trainä¼šè¢«æ”¹å˜ï¼Œå†æ¬¡è¿è¡Œæ—¶æ— æ•ˆçš„    
    
    
    
```

```python
#å¯¹ç±»ç›®åž‹çš„ç‰¹å¾å› å­åŒ–ã€‚ 
#å› å­åŒ–ï¼š
#ä»¥Cabinä¸ºä¾‹ï¼ŒåŽŸæœ¬ä¸€ä¸ªå±žæ€§ç»´åº¦ï¼Œå› ä¸ºå…¶å–å€¼å¯ä»¥æ˜¯[â€˜yesâ€™,â€™noâ€™]ï¼Œè€Œå°†å…¶å¹³å±•å¼€ä¸ºâ€™Cabin_yesâ€™,â€™Cabin_noâ€™ä¸¤ä¸ªå±žæ€§

#åŽŸæœ¬Cabinå–å€¼ä¸ºyesçš„ï¼Œåœ¨æ­¤å¤„çš„â€Cabin_yesâ€ä¸‹å–å€¼ä¸º1ï¼Œåœ¨â€Cabin_noâ€ä¸‹å–å€¼ä¸º0
#åŽŸæœ¬Cabinå–å€¼ä¸ºnoçš„ï¼Œåœ¨æ­¤å¤„çš„â€Cabin_yesâ€ä¸‹å–å€¼ä¸º0ï¼Œåœ¨â€Cabin_noâ€ä¸‹å–å€¼ä¸º1

#å³One-hot é”®

dummies_Cabin = pd.get_dummies(data_train['Cabin'],prefix='Cabin')

dummies_Embarked = pd.get_dummies(data_train['Embarked'], prefix= 'Embarked')

dummies_Sex = pd.get_dummies(data_train['Sex'], prefix= 'Sex')

dummies_Pclass = pd.get_dummies(data_train['Pclass'], prefix= 'Pclass')

df = pd.concat([data_train, dummies_Cabin, dummies_Embarked, dummies_Sex, dummies_Pclass],axis=1)

df.drop(['Pclass','Name','Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)

df
```

```python
#å½’ä¸€åŒ–ï¼Œå°†æ•°å€¼å·®å€¼è¿‡äºŽå¤§çš„æ•°æ®æŒ‰æ¯”ä¾‹ç¼©å°ä»¥è‡³äºŽè®©æ¨¡åž‹æ”¶æ•›ã€‚
#å¯ä»¥ç”¨scikit-learné‡Œé¢çš„preprocessingæ¨¡å—å¯¹è¿™ä¿©å€¼åšä¸€ä¸ªscalingï¼Œæ‰€è°“scalingï¼Œå…¶å®žå°±æ˜¯å°†ä¸€äº›å˜åŒ–å¹…åº¦è¾ƒå¤§çš„ç‰¹å¾åŒ–åˆ°[-1,1]ä¹‹å†…ã€‚

import sklearn.preprocessing as preprocessing
scaler = preprocessing.StandardScaler()
age_data = np.array(df['Age']).reshape(-1,1)
age_scale_param = scaler.fit(age_data)           
df['Age_scaled'] = scaler.fit_transform(age_data,age_scale_param)

fare_data = np.array(df['Fare']).reshape(-1,1)
fare_scale_param = scaler.fit(fare_data)                         
df['Fare_scaled'] = scaler.fit_transform(fare_data,fare_scale_param)
df
                      


```

```python
#é€»è¾‘å›žå½’å»ºæ¨¡
from sklearn import linear_model

#ç”¨æ­£åˆ™å–å‡ºæˆ‘ä»¬è¦çš„å±žæ€§å€¼
train_df = df.filter(regex='Survived|Age_.*|SibSp|Parch|Fare_.*|Cabin_.*|Embarked_.*|Sex_.*|Pclass_.*')
train_np = train_df.as_matrix()

#yå³Survivalç»“æžœ
y = train_np[:,0]

# Xå³ç‰¹å¾å±žæ€§å€¼
X = train_np[:, 1:]

# fitåˆ°RandomForestRegressorä¹‹ä¸­
clf = linear_model.LogisticRegression(C=1.0,penalty='l1',tol=1e-6)
clf.fit(X,y)

clf
```

```python
#å¯¹æµ‹è¯•é›†çš„æ ·æœ¬è¿›è¡ŒåŒæ ·çš„å¤„ç†
data_test = pd.read_csv('test.csv')
data_test.loc[(data_test.Fare.isnull()),'Fare'] = 0
# æŽ¥ç€æˆ‘ä»¬å¯¹test_dataåšå’Œtrain_dataä¸­ä¸€è‡´çš„ç‰¹å¾å˜æ¢
# é¦–å…ˆç”¨åŒæ ·çš„RandomForestRegressoræ¨¡åž‹å¡«ä¸Šä¸¢å¤±çš„å¹´é¾„
tmp_df = data_test[['Age','Fare','Parch','SibSp','Pclass']]
null_age = tmp_df[data_test.Age.isnull()].as_matrix()
#æ ¹æ®ç‰¹å¾å±žæ€§Xé¢„æµ‹å¹´é¾„å¹¶è¡¥ä¸Š
X = null_age[:,1:]

#è¿™é‡Œçš„rfræ˜¯ä¹‹å‰è¿›è¡Œéšæœºç”Ÿæˆçš„æ¨¡åž‹ï¼Œè¦ä½¿ç”¨åŒä¸€ä¸ªæ¨¡åž‹
predictedAges = rfr.predict(X)
data_test.loc[ (data_test.Age.isnull()),'Age' ] = predictedAges

data_test = set_Cabin_type(data_test)
dummies_Cabin = pd.get_dummies(data_test['Cabin'],prefix='Cabin')
dummies_Embarked = pd.get_dummies(data_test['Embarked'],prefix='Embarked')
dummies_Sex = pd.get_dummies(data_test['Sex'],prefix='Sex')
dummies_Pclass =pd.get_dummies(data_test['Pclass'],prefix='Pclass')

df_test = pd.concat([data_test,dummies_Cabin,dummies_Embarked,dummies_Sex,dummies_Pclass],axis=1)
df_test.drop(['Pclass','Name', 'Sex', 'Ticket', 'Cabin', 'Embarked'], axis=1, inplace=True)

age_test_data = np.array(df_test['Age']).reshape(-1,1)
age_scale_param = scaler.fit(age_test_data)           
df_test['Age_scaled'] = scaler.fit_transform(age_test_data,age_scale_param)

fare_test_data = np.array(df_test['Fare']).reshape(-1,1)
fare_scale_param = scaler.fit(fare_test_data)                         
df_test['Fare_scaled'] = scaler.fit_transform(fare_test_data,fare_scale_param)

df_test
```
